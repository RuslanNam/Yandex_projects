## 2. Обучить и проверить модели для каждого региона:
На данном этапе стоит задача потсроения модели для предсказания объёма запасов в скважинах по линейной регрессии

### 2.1. Разбить данные на обучающую и валидационную выборки в соотношении 75:25.
Выделим признаки и целевой признак по каждой из датасетов, а затем поделим данные на обучающие(по 75%) и валидационные(по 25%)

```python
#data0:
features0 = data0.drop(['id','product'],axis=1)
target0 = data0['product']
features_train0, features_valid0, target_train0, target_valid0 = train_test_split(features0, target0 ,test_size=0.25, random_state=12)

#data1:
features1 = data1.drop(['id','product'],axis=1)
target1 = data1['product']
features_train1, features_valid1, target_train1, target_valid1 = train_test_split(features1, target1 ,test_size=0.25, random_state=12)

#data2:
features2 = data2.drop(['id','product'],axis=1)
target2 = data2['product']
features_train2, features_valid2, target_train2, target_valid2 = train_test_split(features2, target2 ,test_size=0.25, random_state=12)
```

### 2.2. Обучить модель и сделать предсказания на валидационной выборке.

### 2.3. Сохранить предсказания и правильные ответы на валидационной выборке.


### 2.4. Напечатать на экране средний запас предсказанного сырья и RMSE модели.

```python
#Запишем предсказания по регионам в следующие переменные:
predictions0 = []
predictions1 = []
predictions2 = []
# Создадим списки из нужных переменных для цикла
preds = [predictions0, predictions1, predictions2]

features_train = [features_train0, features_train1, features_train2]
features_valid = [features_valid0, features_valid1, features_valid2]
target_train = [target_train0, target_train1, target_train2]
target_valid = [target_valid0, target_valid1, target_valid2]

for i in range(len(features_train)):
    model = LinearRegression()
    model.fit(features_train[i], target_train[i])
    predictions = model.predict(features_valid[i])
    for n in predictions:
        if n not in preds[i]:
            preds[i].append(n)
    print('region',i)
    print('Средний запас предсказанного сырья:',predictions.mean())
    print('Фактический средний запас сырья:',target_valid[i].mean())
    print('RMSE модели на валидационной выборке:', mean_squared_error(target_valid[i], predictions)**0.5)
```

## 4. Написать функцию для расчёта прибыли по выбранным скважинам и предсказаниям модели:

### 4.1. Выбрать скважины с максимальными значениями предсказаний.

### 4.2. Просуммировать целевое значение объёма сырья, соответствующее этим предсказаниям.

### 4.3. Рассчитать прибыль для полученного объёма сырья.
Функция должна выбрать 200 лучших скважин и на основе ответов подсчитать возможную прибыль.

```python
predictions0 = pd.Series(predictions0, index=target_valid0.index)
predictions1 = pd.Series(predictions1, index=target_valid1.index)
predictions2 = pd.Series(predictions2, index=target_valid2.index)
```

```python
def profit(target, prediсtions, n, COSTS):
    # Выбираем скважины с максимальными значениями предсказаний
    sorted_prediсtions = prediсtions.sort_values(ascending=False)
    # Выбираем из target индексы от predictions
    selected_target = target[sorted_prediсtions.index][:n]
    # Суммируем целевое значение объёма сырья
    quantity = selected_target.sum()
    # Считаем выручку, average_revenue = 450 тыс.руб
    revenue = quantity*AVG_REVENUE
    # Считаем расходы за все 200 лучших скважин
    COSTS = COSTS*200
    # Считаем прибыль
    profit = revenue - COSTS
    return profit      
```

<div class="alert alert-block alert-success">
<b>Успех:</b> Верно.
</div>

```python
profit0 = profit(target_valid0, predictions0, 200, COSTS)
profit1 = profit(target_valid1, predictions1, 200, COSTS)
profit2 = profit(target_valid2, predictions2, 200, COSTS)

print('Прибыль с region 0 = {:.3f} тыс. рублей'.format(profit0))
print('Прибыль с region 1 = {:.3f} тыс. рублей'.format(profit1))
print('Прибыль с region 2 = {:.3f} тыс. рублей'.format(profit2))
```

## 5. Посчитайте риски и прибыль для каждого региона:

### 5.1. Примените технику Bootstrap с 1000 выборок, чтобы найти распределение прибыли.

### 5.2. Найдите среднюю прибыль, 95%-й доверительный интервал и риск убытков. Убыток — это отрицательная прибыль.

### 5.3. Напишите выводы: предложите регион для разработки скважин и обоснуйте выбор.

```python
state = np.random.RandomState(12)
target_valids = [target_valid0, target_valid1, target_valid2]
predictions = [predictions0, predictions1, predictions2]

#region 0 
values0 = []
neg_values0 = 0
for i in range(1000):
    target_subsample = target_valids[0].sample(n=500, replace=True, random_state=state)
    pred_subsample = predictions[0][target_subsample.index]
    res = profit(target_subsample, pred_subsample, 200, COSTS)
    values0.append(res)
    if res < 0:
        neg_values0 += 1

values0 = pd.Series(values0)
mean0 = values0.mean()
lower0 = values0.quantile(0.025)
upper0 = values0.quantile(0.975)

print("Средняя прибыль region 0:", mean0)
print('95% доверительный интервал: ({:.3f}, {:.3f})'.format(lower0, upper0))
print('Риск убытков:', neg_values0/len(values0))
print('-----------------------')

#region 1
values1 = []
neg_values1 = 0

for i in range(1000):
    target_subsample = target_valids[1].sample(n=500, replace=True, random_state=state)
    pred_subsample = predictions[1][target_subsample.index]
    res = profit(target_subsample, pred_subsample, 200, COSTS)
    values1.append(res)
    if res < 0:
        neg_values1 += 1

values1 = pd.Series(values1)
mean1 = values1.mean()
lower1 = values1.quantile(0.025)
upper1 = values1.quantile(0.975)

print("Средняя прибыль region 1:", mean1)
print('95% доверительный интервал: ({:.3f}, {:.3f})'.format(lower1, upper1))
print('Риск убытков:', neg_values1/len(values1))
print('-----------------------')

#region 2
values2 = []
neg_values2 = 0

for i in range(1000):
    target_subsample = target_valids[2].sample(n=500, replace=True, random_state=state)
    pred_subsample = predictions[2][target_subsample.index]
    res = profit(target_subsample, pred_subsample, 200, COSTS)
    values2.append(res)
    if res < 0:
        neg_values2 += 1

values2 = pd.Series(values2)
mean2 = values2.mean()
lower2 = values2.quantile(0.025)
upper2 = values2.quantile(0.975)

print("Средняя прибыль region 2:", mean2)
print('95% доверительный интервал: ({:.3f}, {:.3f})'.format(lower2, upper2))
print('Риск убытков:', neg_values2/len(values2))
print('-----------------------')
```
